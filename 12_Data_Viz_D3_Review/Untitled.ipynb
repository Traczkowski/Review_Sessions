## High level data viz and dashboarding concepts

- Today, I wanted to share a lot of great resources with you all, spend a little bit of time talking about dashboarding and data visualization in practirce, and bring forward some things that have inspired me over the years that I hope can help you all as well! Since last week was a little bit more code and demo-heavy, this week we are going to step back a little bit and talk about data viz at a high level.

#### Hans Rosling (Physician, but mainly his TED Talk on stats)
- https://www.ted.com/talks/hans_rosling_the_best_stats_you_ve_ever_seen?language=en#t-280606
- Hans Rosling on showing data in a meaningful way and how useful it is. One of the most memorable Ted talks I've ever seen and an inspiration to me years before I thought I'd be getting into this field. Though he was a doctor the excitement with which he spoke about data visualization was incredibly inspiring and I'm sure he's inspired countless others to think about data and how to show it to others in a way that matters and could drive change.

#### The Atlas of Economic Complexity (Harvard Research Project)
- https://atlas.cid.harvard.edu/
- http://globe.cid.harvard.edu/ 
- The Atlas of Economic complexity is something that has fascinated me and is a fantastic resource that shows so many different types of amazing visualizations under one roof, from a standard tree plot, to a 3 dimensional globe that I will barely be able to share with you because it brings my computer to its knees even when I'm not streaming.

#### The Python Graph Gallery (Website)
- https://python-graph-gallery.com/
- I know that I've pushed the Python Graph Gallery before, and they not only have pre-built code for a ton of fantastic visualizations, but they also have some very nice notes and heuristics on what you should and shouldn't do with charts, I wanted to talk a little bit about the "bad chart" section today just so that I do my due dilligence to share some things to stay away from.

#### XKCD (Webcomic)
- https://xkcd.com/
- http://www.vislives.com/2011/10/xkcd-visualizations.html
- https://xkcd.com/688/
- XKCD is a webcomic that some of you may be familiar with. The content ranges from a few panels with some stick figures in it, to incredibly splashy visualizations with an insane amount of detail in them. The cartoonist, Randall Munroe, is an engineer by trade and has been drawing cartoons since I was in gradeschool. As I understand it he draws his cartoons by hand, but he has had quite a bit of influence on many people in the world of math/computer science. If you think I'm kidding, feel free to check out the XKCD style plots that are available in Matplotlib(!) and Seaborn (http://alimanfoo.github.io/2016/05/31/matplotlib-xkcd.html). A lot of his visualizations will have bits that are tongue in cheek or serve to poke fun, but the level of detail of his visualizations are incredible and he makes it easy to have fun while learning a thing or two. This is another one person whose work I've been enjoying for years before I realized that it was inspiring.


#### 3Blue1Brown (Youtube Channel)
- https://www.youtube.com/watch?v=aircAruvnKk&list=PLZHQObOWTQDNU6R1_67000Dx_ZCJB-3pi
- https://www.youtube.com/watch?v=IHZwWFHWa-w&t=822s
- 3Blue1Brown's (Grant Sanderson) work has been invaluable to me, and that is specifically the visualizations that he brings to the table when talking about a lot of math (and some programming) concepts. I have to say that I was someone whose eyes would glaze over in classes when my professors would talk about math in 2, 3 or N dimensions and there would be absolutely nothing to look at to try to understand it. They would maybe throw up 3 or 4 lines on the whiteboard or rotate their wrist in a way to show that "Yes this does exist in space!", but that wasn't really doing it for me. The focus here is to show you what the math is doing, I watched this guy's work constantly when I was in my own bootcamp, and I think that the way that he painstakingly animates the math and tries to describe it while using the viz as a core part of his teaching philosophy can help so many people get a better understanding of math concepts that they'd typically reel away from. Although we don't really explore the video format in here, Grant puts on a masterclass in using visualization to bring knowledge to more people.

#### r/dataisbeautiful and r/dataisugly (Subreddits)
- https://www.reddit.com/r/dataisbeautiful/
- https://www.reddit.com/r/dataisugly/
- Although the content varies day to day and sometimes it is hit-or-miss, you will see a large variety of different types of viz from the core groups (parts of a whole, trend over time, maps, etc.) that are constantly rotating with trends. Recently, the dominant topic has been Covid, but you can always pop over there, scroll around for a little bit, and see both good and bad examples along with strong opinions in either direction (EX: "This graph is awful because X! How could you have done Y?!") This can be a great inspiration if you have similar data available to you. Datasources are often shared on there as well so you can often see what somebody did, pick up that same data, and see what you can come up with yourself if you're looking for a challenge.

#### Storytelling With Data (Book and Blog)
- http://www.storytellingwithdata.com/
- This book is not one that I've ever sat down and read cover-to-cover, but one that I've found myself going back to several times over the last few years for some best-practices. It opens with a quote: "Power Corrupts. PowerPoint Corrupts Absolutely" which I get a kick out of, and it speaks to the rest of the book which tries to cut through a lot of the flashy stuff that you may see out in the field and tries to strike at the heart of good, clean viz. I'll share some chapter titles...
